```python
X_train = training_set.iloc[:, list(range(0,20)) + list(range(21,39))] # removing the label column
y_train = training_set.loc[:, "label"]

X_test = test_set.iloc[:, list(range(0,20)) + list(range(21,39))]
y_test = test_set.loc[:, "label"]
#IoTdataset.info()
print(X_train.info(), y_train)
print("X train:", type(X_train))
print("y train:", type(y_train))

print(X_test.info(), y_test)
print("X test:", type(X_test))
print("y test:", type(y_test))
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
y_train_encoded = encoder.fit_transform(y_train)
y_test_encoded = encoder.transform(y_test)

print(encoder.classes_) # command to find the mapping between the initial labels and the new numerical classes
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
import pickle


my_pipeline = Pipeline( [ ("scaler", StandardScaler()), ("rf_clf", RandomForestClassifier()) ] )

param_grid = {'rf_clf__max_depth': [2, 5, 10, 100], 'rf_clf__n_estimators': [5, 10, 20]}
grid_search = GridSearchCV(my_pipeline, param_grid, cv=4, scoring='neg_mean_squared_error')

grid_search.fit(X_train, y_train_encoded)
clf = grid_search.best_estimator_

pickle.dump(clf, open('my_model.sav', 'wb'))
```

```python
loaded_model = pickle.load(open('my_model.sav', 'rb'))

print("best score on training:", grid_search.best_score_)
print("train accuracy: ", accuracy_score(y_train_encoded, loaded_model.predict(X_train)))

y_pred = loaded_model.predict(X_test)
print("test accuracy: ", accuracy_score(y_test_encoded, y_pred))
print("test precision: ", precision_score(y_test_encoded, y_pred, average=None))
print("test micro-average precision: ", precision_score(y_test_encoded, y_pred, average="micro"))
print("test recall: ", recall_score(y_test_encoded, y_pred, average=None))
print("test micro-average recall: ", recall_score(y_test_encoded, y_pred, average="micro"))
print("y pred: ", y_pred)
```
